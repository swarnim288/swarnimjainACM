```python
import numpy as np
import pandas as pd
import matplotlib as mt
```

```python
data=pd.read_csv('best-fit.csv')
```

```python
x= data['X']
y=data['Y']
slope,intercept = np.polyfit(x,y,1)
bf1=slope*x+intercept
mt.pyplot.scatter(x,y, label='data points')
mt.pyplot.plot(x,bf1,color='red' , label='bestfit line')
mt.pyplot.xlabel('X')
mt.pyplot.ylabel('Y')
mt.pyplot.legend()
mt.pyplot.show()
```
